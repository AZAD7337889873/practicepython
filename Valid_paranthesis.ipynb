```python
def is_valid_parentheses(s: str) -> bool:
    # Map closing brackets to their corresponding opening brackets
    bracket_map = {')': '(', '}': '{', ']': '['}
    stack = []

    for char in s:
        # If it's a closing bracket
        if char in bracket_map:
            # Pop the top element if stack isn't empty; else use a dummy variable
            top_element = stack.pop() if stack else '#'
            
            # If it doesn't match the required opening bracket, it's invalid
            if bracket_map[char] != top_element:
                return False
        else:
            # It's an opening bracket, push it onto the stack
            stack.append(char)
            
    # If the stack is empty, all brackets were matched perfectly
    return len(stack) == 0

# --- Corrected Test Cases ---
test_strings = ["()", "()[]{}", "(]", "([)]", "{[]}", "((( Grimm )"]

for index, test in enumerate(test_strings, 1):
    print(f"Test {index}: '{test}' -> {is_valid_parentheses(test)}")
```

```python
def is_valid_parenthesis(s: str) -> bool:
    # Intuitive mapping: Opening bracket -> Closing bracket
    bracket = {'(': ')', '[': ']', '{': '}'}
    stack = []
    
    for char in s:
        # If it's an opening bracket, push it to the stack
        if char in bracket:
            stack.append(char)
        else:
            # It's a closing bracket. If stack is empty, there is no opening partner.
            if not stack:
                return False
            
            # Pop the last opening bracket and check if its expected 
            # closing bracket matches the current character
            top_element = stack.pop()
            if bracket[top_element] != char:
                return False
                
    # If the stack is empty, all brackets were perfectly matched
    return len(stack) == 0

# --- Test Cases ---
test_strings = ['()', '[]', '{()}', '([)]', '(]', '(', ']']
for index, test in enumerate(test_strings, 1):
    print(f"Test {index}: '{test:<5}' -> {is_valid_parenthesis(test)}")
```

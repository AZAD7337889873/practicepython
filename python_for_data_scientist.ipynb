```python
# Given an array, find all the duplicates in this array? For example: input: [1,2,3,1,3,6,5] output: [1,3]
```

```python
def duppli(nums):
    seen = set()
    duplicate = set()
    for num in nums:
        if num in seen:
            duplicate.add(num)
        else:
            seen.add(num)
            
    return list(duplicate)
```

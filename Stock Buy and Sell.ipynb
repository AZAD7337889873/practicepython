```python
def max_profit(prices):
    n = len(prices)
    res =0
    for i in range(n-1):
        for j in range(i+1, n):
            res = max(res, prices[j]-prices[i])
    return res

```

```python
def max_profit(prices):
    n = len(prices)
    minsofar = prices[0]
    res = 0
    for i in range(1,n):
        minsofar = min(minsofar, prices[i])
        res = max(res, prices[i]-minsofar)
    return res
```

```python
print(max_profit(([1,2,3,4,543,23,4424,5,4353])))
```

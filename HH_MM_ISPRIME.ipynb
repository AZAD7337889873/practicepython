```python
def prime_time() -> list[int]:
    primes = []
    def isprime(n):
        if n<2:
            return False
        for i in range(2,int(n**0.5)):
            if n%i ==0:
                return False
        return True
    for h in range(24):
        for m in range(60):
            hhmm = h*100+m
            if isprime(hhmm):
                primes.append(hhmm)
    return primes
```

```python
if __name__ == "__main__":
    result = prime_time()
    print(*result)
```

```python
def prime_t() -> list[int]:
    p = []
    def is_prime(n):
        if n<2:
            return False
        for i in range(2, int(n**0.5)):
            if n%i == 0:
                return False
        return True
    for h in range(24):
        for m in range(60):
            hhmm = h*100+m
            if is_prime(hhmm):
                p.append(hhmm)
    return p

if __name__ == "__main__":
    result = prime_t()
    print(" ".join(map(str,result)))
```

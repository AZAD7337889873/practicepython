```python
roots = ['cat', 'bat','rat']
sentence = 'the cattle was rattled by the battery'

def replace_word(roots, sentence):
    words = sentence.split(" ")
    for index, word in enumerate(words):
        for root in  roots:
            if word.startswith(root):
                words[index] = root
                break
    return " ".join(words)

replace_word(roots,sentence)
```

```python
roots = ['rat', 'cat', 'bat']
sentence = 'the cattle was rattled by the battery'

def replace_roots(roots, sentence):
    words = sentence.split(" ")
    for index, word in enumerate(words):
        for root in roots:
            if word.startswith(root):
                words[index] = root
    return  " ".join(words)

replace_roots(roots, sentence)
```

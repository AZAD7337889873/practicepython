```python
def solution(s):
    count = [0]*10
    for char in s:
        count[int(char)]  +=1 
    left_part = []
    for digit in range(9,-1,-1):
        if digit == 0 and len(left_part) == 0:
            continue
        pairs = count[digit]//2
        if pairs > 0:
            left_part.append(str(digit)*pairs)
            count[digit] -= pairs*2
            
    left_str = "".join(left_part)
    middle_digit = ""
    for digit in range(9,-1,-1):
        if count[digit]>0:
            middle_digit = str(digit)
            break
        
    Result = left_str + middle_digit + left_str[::-1]
    if not Result:
        return "0"
    return Result
```

```python
# Copy and paste your input like this:
print(solution("4343212"))  # Output: 4321234
print(solution("13579"))    # Output: 9
print(solution("000"))      # Output: 0
```
